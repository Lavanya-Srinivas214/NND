```python
import tensorflow as tf
x = tf.constant([0.5, 0.2, 0.1], dtype=tf.float32)
w = tf.Variable([[0.1, 0.4],
                 [0.2, 0.3],
                 [0.3, 0.6]], dtype=tf.float32)
b = tf.Variable([0.01, 0.02], dtype=tf.float32)
z = tf.matmul(tf.expand_dims(x, 0), w) + b
a = tf.nn.sigmoid(z)
print("Output z:", z.numpy())
print("Activation Output a:", a.numpy())
w2 = tf.Variable([0.2, 0.3], dtype=tf.float32)
b2 = tf.Variable([0.05], dtype=tf.float32)
y = tf.constant([1.0], dtype=tf.float32)
lr = 0.1
with tf.GradientTape() as tape:
    z2 = tf.matmul(a, tf.expand_dims(w2, 1)) + b2
    a2 = tf.nn.sigmoid(z2)
    loss = tf.square(y - a2)
grads = tape.gradient(loss, [w2, b2])
w2.assign_sub(lr * grads[0])
b2.assign_sub(lr * grads[1])
print("Final output:", a2.numpy())
print("Loss:", loss.numpy())
print("Updated w2:", w2.numpy())
print("Updated b2:", b2.numpy())
```
