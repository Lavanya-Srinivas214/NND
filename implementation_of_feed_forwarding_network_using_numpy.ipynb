```python
x=np.array([0.5,0.2,0.1])
w=np.array([[0.1,0.4],
           [0.2,0.3],
           [0.3,0.6]])
b=np.array([0.01,0.02])
z=np.dot(x,w)+b
print("Output",z)
def sigmoid(x):
    return 1/(1+np.exp(-x))
a=sigmoid(z)
print("Activation Output",a)
w2=np.array([0.2,0.3])
b2=np.array([0.05])
z2=np.dot(a,w2)+b2
a2=sigmoid(z2)
print("Final output:",a2)
y=1
loss=(y-a2)**2
print("Loss:",loss)
lr=0.1
dloss_de2=-2+(y-a2)
de2_dz2=a
dz2_dw2=a # Defining dz2_dw2
grad_w2 = dloss_de2*de2_dz2*dz2_dw2
w2=w2-lr*grad_w2.reshape(2,1)
print("updated w2",w2)
```
