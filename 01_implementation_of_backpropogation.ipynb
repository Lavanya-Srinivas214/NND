```python
import numpy as np

# Input and output
X = np.array([[0,0],[0,1],[1,0],[1,1]])
y = np.array([[0],[1],[1],[0]])  # XOR

# Initialize weights
np.random.seed(1)
W1 = np.random.randn(2, 3)
W2 = np.random.randn(3, 1)
lr = 0.1

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def sigmoid_deriv(x):
    return x * (1 - x)

# Training
for epoch in range(5000):
    # Forward pass
    z1 = sigmoid(np.dot(X, W1))
    output = sigmoid(np.dot(z1, W2))

    # Error
    error = y - output

    # Backpropagation
    d_output = error * sigmoid_deriv(output)
    d_hidden = np.dot(d_output, W2.T) * sigmoid_deriv(z1)

    # Update weights
    W2 += np.dot(z1.T, d_output) * lr
    W1 += np.dot(X.T, d_hidden) * lr

# Final output
print("Predicted Output:\n", output.round(3))
```

```python
import tensorflow as tf
import numpy as np

# Input and output
X = np.array([[0,0],[0,1],[1,0],[1,1]])
y = np.array([[0],[1],[1],[0]])

# Build model
model = tf.keras.Sequential([
    tf.keras.layers.Dense(3, activation='sigmoid', input_shape=(2,)),
    tf.keras.layers.Dense(1, activation='sigmoid')
])

# Compile model
model.compile(optimizer='adam', loss='binary_crossentropy')

# Train model
model.fit(X, y, epochs=2000, verbose=0)

# Predict
pred = model.predict(X)
print("Predicted Output:\n", np.round(pred, 3))
```
